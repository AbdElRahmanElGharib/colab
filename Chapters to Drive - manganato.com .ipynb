```python
import sys
import time
import os
import cv2
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from PIL import Image
from google.colab import drive
from google.colab import files
import urllib.request
from pyvirtualdisplay import Display

display = Display(visible=False, size=(1920, 1080), use_xauth=True)
display.start()

import pyautogui

sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

PAGE_IMAGE_XPATH = '/html/body/div[1]/div[3]/img'
CHAPTER_LIST_XPATH = '/html/body/div[1]/div[3]/div[1]/div[3]/ul/li/a'
LIST_EXTENDER_XPATH = '//*[@id="panel-story-info-description-show-more"]'
TITLE_XPATH = '/html/body/div[1]/div[3]/div[1]/div[2]/div[2]/h1'

urllib.request.urlretrieve('https://drive.google.com/uc?id=1uxirWbpar6j_jPsNWFtagpXeNKbBZxBK&export=download', 'context_menu.png')

drive.mount('/content/drive')

class DownloadChapter:

    download_path = '/content/'

    names = []

    def __init__(self, pdf_name, target_url, save_path):
        chromedriver_path = '/usr/lib/chromium-browser/chromedriver'

        #image XPATH
        target_xpath = PAGE_IMAGE_XPATH

        chrome_options = Options()
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')

        driver = webdriver.Chrome(service=Service(chromedriver_path),
                                  desired_capabilities=DesiredCapabilities.CHROME.copy(),
                                  options=chrome_options)
        desired_action = ActionChains(driver)

        driver.get(target_url)

        time.sleep(5)

        elements = driver.find_elements(by=By.XPATH, value=target_xpath)

        start_image = Image.Image()
        images = []

        for element in elements:
            src = element.get_attribute('src')
            name = src[src.rfind('/') + 1:]
            self.names.append(name)

            desired_action.move_to_element(to_element=element).context_click().perform()
            desired_action.reset_actions()

            x, y = pyautogui.locateCenterOnScreen('context_menu.png', confidence=0.9)
            pyautogui.moveTo(x, y)
            pyautogui.leftClick()

            pyautogui.moveTo(x=100, y=30)
            pyautogui.leftClick()

            with pyautogui.hold('ctrl'):
                pyautogui.press('end')
            with pyautogui.hold('shift'):
                with pyautogui.hold('ctrl'):
                    pyautogui.press('home')
            pyautogui.press('backspace')

            pyautogui.write(name)

            pyautogui.moveTo(x=50, y=155)
            pyautogui.doubleClick()

            pyautogui.moveTo(x=1080, y=800, duration=0.6)
            pyautogui.leftClick()

        time.sleep(1)

        for i in range(len(elements)):
            if i == 0:
                try:
                    start_image = Image.open(self.download_path + self.names[i])

                except FileNotFoundError:
                    desired_action.move_to_element(to_element=elements[i]).context_click().perform()
                    desired_action.reset_actions()

                    x, y = pyautogui.locateCenterOnScreen('context_menu.png', confidence=0.9)
                    pyautogui.moveTo(x, y)
                    pyautogui.leftClick()

                    pyautogui.moveTo(x=100, y=30)
                    pyautogui.leftClick()
            
                    with pyautogui.hold('ctrl'):
                        pyautogui.press('end')
                    with pyautogui.hold('shift'):
                        with pyautogui.hold('ctrl'):
                            pyautogui.press('home')
                    pyautogui.press('backspace')

                    pyautogui.write(self.names[i])

                    pyautogui.moveTo(x=50, y=155)
                    pyautogui.doubleClick()

                    pyautogui.moveTo(x=1080, y=800, duration=0.6)
                    pyautogui.leftClick()

                    time.sleep(1)

                    start_image = Image.open(self.download_path + self.names[i])

                except Exception as e:
                    raise e

            else:
                try:
                    images.append(Image.open(self.download_path + self.names[i]))

                except FileNotFoundError:
                    desired_action.move_to_element(to_element=elements[i]).context_click().perform()
                    desired_action.reset_actions()

                    x, y = pyautogui.locateCenterOnScreen('context_menu.png', confidence=0.9)
                    pyautogui.moveTo(x, y)
                    pyautogui.leftClick()

                    pyautogui.moveTo(x=100, y=30)
                    pyautogui.leftClick()
            
                    with pyautogui.hold('ctrl'):
                        pyautogui.press('end')
                    with pyautogui.hold('shift'):
                        with pyautogui.hold('ctrl'):
                            pyautogui.press('home')
                    pyautogui.press('backspace')

                    pyautogui.write(self.names[i])

                    pyautogui.moveTo(x=50, y=155)
                    pyautogui.doubleClick()

                    pyautogui.moveTo(x=1080, y=800, duration=0.6)
                    pyautogui.leftClick()

                    time.sleep(1)

                    images.append(Image.open(self.download_path + self.names[i]))

                except Exception as e:
                    raise e

        start_image.save(save_path + pdf_name + '.pdf', "PDF", resolution=100.0, save_all=True, append_images=images)

        time.sleep(0.1)

        driver.quit()

    def __del__(self):

        for name in self.names:
            file_path = '/content/' + name
            if os.path.exists(file_path):
                os.remove(file_path)


class DownloadManhwa:

    chapters_links = []
    chapters_names = []
    title = ''

    def __init__(self, manhwa_url):

        chromedriver_path = '/usr/lib/chromium-browser/chromedriver'

        #chapters list item XPATH
        target_xpath = CHAPTER_LIST_XPATH

        #chapters list etender button XPATH
        button_xpath = LIST_EXTENDER_XPATH

        #manhwa title XPATH
        title_xpath = TITLE_XPATH

        chrome_options = Options()
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')

        driver = webdriver.Chrome(service=Service(chromedriver_path),
                                  desired_capabilities=DesiredCapabilities.CHROME.copy(),
                                  options=chrome_options)

        driver.get(manhwa_url)
        
        try:
            step_action = ActionChains(driver)
            step_action.move_to_element(driver.find_element(by=By.XPATH, value=button_xpath)).click()
            step_action.perform()
            step_action.reset_actions()
        except:
            pass

        elements = driver.find_elements(by=By.XPATH, value=target_xpath)

        for element in elements:
            link = element.get_attribute('href')
            self.chapters_links.append(link)
            self.chapters_names.append(element.text)
        self.chapters_names.reverse()
        self.chapters_links.reverse()

        element = driver.find_element(by=By.XPATH, value=title_xpath)

        self.title = element.text

        driver.quit()

        time.sleep(0.3)
        for i in range(len(self.chapters_names)):
            print(i, '       ', self.chapters_names[i])

        print('\n')

    def get_title(self) -> str:
      return self.title
    
    def download(self, path, start_index, end_index):
        for i in range(start_index, end_index+1):
            print('downloading:    ', self.chapters_names[i])

            chapter_downloader = DownloadChapter(pdf_name=self.chapters_names[i],
                                                 target_url=self.chapters_links[i],
                                                 save_path=path)
            del chapter_downloader
            time.sleep(0.25)

input_save_path = '/content/drive/MyDrive/manhwa/'

if not os.path.exists(input_save_path):
    os.mkdir(input_save_path)

print('\n\n\n')

key = 'y'
while True:
    if key == 'y' or key == 'Y':
        start_time = time.time()

        mode = input('Do you want to download a chapter or the whole manhwa?\n'
                     '      for One Chapter type [C]    and for the Whole Manhwa type [M]\n')
        if mode == 'm' or mode == 'M':
            input_target_url = input('Enter Manhwa URL:\n')
            
            start_time = time.time()

            downloader = DownloadManhwa(manhwa_url=input_target_url)

            manhwa_name = downloader.get_title()

            input_save_path += manhwa_name + '/'
            if not os.path.exists(input_save_path):
                os.mkdir(input_save_path)

            print('Download chapters from index:\n')
            a = int(input())
            print('to index:\n')
            b = int(input())

            print('\n\n')

            downloader.download(input_save_path, a, b)
            del downloader

            print("--- %s seconds ---" % (time.time() - start_time))

        elif mode == 'c' or mode == 'C':
            input_target_url = input('Enter Chapter URL:\n')
            input_pdf_name = input('Enter file name to be saved (without extensions):\n')

            start_time = time.time()

            print('downloading:  ', input_pdf_name)

            downloader = DownloadChapter(save_path=input_save_path,
                                         target_url=input_target_url,
                                         pdf_name=input_pdf_name)
            del downloader

            print("--- %s seconds ---" % (time.time() - start_time))

        else:
            print('\n                 INVALID INPUT                   \n\n')

    else:
        break

    key = input('Do  you want download another content?\n'
                '       Enter [Y] to continue   or   Enter anything else to exit:\n')

display.stop()

if os.path.exists('/content/context_menu.png'):
    os.remove('/content/context_menu.png')
```
